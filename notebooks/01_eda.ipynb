## üì¶ Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

sns.set(style="whitegrid")
```

## üìÅ Load Dataset

```python
df = pd.read_csv("../data/insurance_synthetic.csv")
df.head()
```

## üîç Identify Numerical & Categorical Columns

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
# Auto-detect date columns
date_cols = [
    col for col in categorical_cols
    if pd.to_datetime(df[col], format="%d-%m-%Y", errors='coerce').notna().sum() > 0
]
# Remove date_cols from categorical_cols
categorical_cols = [col for col in categorical_cols if col not in date_cols]

print("Numerical Columns:\n", numerical_cols)
print("Categorical Columns:\n", categorical_cols)
print("Date Columns:\n", date_cols)
```

## üß© Feature Groups (Based on Dataset Schema)

This dataset contains the following logical feature groups:

- **Demographics:** Age, Gender, Marital Status, Education Level  
- **Socioeconomic:** Occupation, Income Level, Credit Score  
- **Geographic:** Geographic Information, Location  
- **Behavioral:** Behavioral Data, Purchase History, Interactions with Customer Service  
- **Policy Details:** Policy Start Date, Policy Renewal Date, Policy Type  
- **Financial:** Coverage Amount, Premium Amount, Deductible  
- **Risk Indicators:** Risk Profile, Driving Record, Previous Claims History  
- **Preferences:** Customer Preferences, Preferred Communication Channel, Preferred Contact Time, Preferred Language  
- **Segmentation:** Segmentation Group (target for clustering)

## üìà Numerical Feature Distributions

```python
rows = math.ceil(len(numerical_cols) / 2)

fig, axes = plt.subplots(rows, 2, figsize=(12, 4 * rows))
axes = axes.flatten() # convert 2D to 1D array

for idx, col in enumerate(numerical_cols):
    sns.histplot(df[col], kde=True, ax=axes[idx], color="teal")
    axes[idx].set_title(f"Distribution of {col}")

# Hide unused subplots if odd number of date columns
for j in range(idx + 1, len(axes)):
    fig.delaxes(axes[j])
    
plt.tight_layout()
plt.show()
```

## üß© Categorical Feature Distributions

```python
rows = math.ceil(len(categorical_cols) / 2)

fig, axes = plt.subplots(rows, 2, figsize=(16, 4 * rows))
axes = axes.flatten()

for idx, col in enumerate(categorical_cols):
    sns.countplot(x=df[col], ax=axes[idx], color="teal")
    axes[idx].set_title(f"Distribution of {col}")
    axes[idx].tick_params(axis='x', rotation=90)

# Hide unused subplots if odd number of date columns
for j in range(idx + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
rows = math.ceil(len(date_cols) / 2)
fig, axes = plt.subplots(rows, 2, figsize=(14, 4 * rows))
axes = axes.flatten()

for col in date_cols:
    df[col] = pd.to_datetime(df[col].astype(str), format="%d-%m-%Y", errors="coerce")

for idx, col in enumerate(date_cols):

    # Group by month for smoother trend
    monthly_counts = df.groupby(df[col].dt.to_period('M')).size()

    # Plot
    axes[idx].plot(
        monthly_counts.index.astype(str),
        monthly_counts.values,
        marker='o',
        color='teal'
    )

    axes[idx].set_title(f"Trend Over Time: {col}")
    axes[idx].set_xlabel("Month")
    axes[idx].set_ylabel("Count")
    axes[idx].tick_params(axis='x', rotation=90)

# Hide unused subplots if odd number of date columns
for j in range(idx + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## üî• Correlation Heatmap (Numerical Features)

```python
plt.figure(figsize=(14,10))
sns.heatmap(df[numerical_cols].corr(), cmap="BuGn", annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## üéØ Segmentation Group Analysis

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Segmentation Group", data=df, color="teal")
plt.title("Distribution of Segmentation Groups")
plt.show()
```
